### Node visualization for subsets of activations found across dimensions within the last layer

This notebook shows the code steps to:
- Extract the most common salient activations found across 100 runs.
- UpSet plots and Venn diagrams to see the intersection of activations across different topics.

```python
from utils.utils_nodes import most_common_nodes
from utils.utils_viz import plot_upset_politics, plot_upset_ethical
import venn
import warnings

warnings.filterwarnings("ignore")
```

##### Visualization for five ethical theories

```python
output_files_dict_ethics = {
    "RELAT": "../output/adv_output_subscribestoculturalrelativism.out",
    "DEONT": "../output/adv_output_subscribestodeontology.out",
    "NIHIL": "../output/adv_output_subscribestomoralnihilism.out",
    "UTILI": "../output/adv_output_subscribestoutilitarianism.out",
    "VIRTU": "../output/adv_output_subscribestovirtueethics.out",
}

ethics_topics_nodes = most_common_nodes(output_files_dict_ethics)
plot_upset_ethical(ethics_topics_nodes)
```

```python
labels = venn.get_labels(ethics_topics_nodes.values(), fill=["number"])
fig, ax = venn.venn5(labels, names=ethics_topics_nodes.keys())
fig.show()
```
